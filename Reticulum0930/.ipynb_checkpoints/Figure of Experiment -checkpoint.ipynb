# The simulation of Gearbox

1. The overlap and size of Gearbox

```python
import random

"""The simulation about Gearbox:
    We generated 2400 nodes and there are five gears, gears size are 18,29,38,46,221. We calculate it from the Shard size.ipynb with failure rate fall in 10^-7 and 10^-8.
    We assign nodes into shard randomly and check if the adversarial nodes in the shard is more than this shard's livness threshold or we turn to a bigger one.
    we keep doing this until all the shard can run normally and then calculate the average shard size. Also,we calculate the overlap time.
"""

"""Function run_simulation used for find the average shard size of Gearbox in different adversarial nodes ratio.
    Input:
        n: The adversarial nodes ratio in the system(upper bound).
        
    Output:
        average shard size of all shard
        
"""



def run_simulation(n):
    # Calculate the number of Byzantine nodes based on the percentage 'n'
    byzantine_nodes = int(n / 100 * 2400)
    results = []
    for i in range(15):
        # Step 1: Select a sample of 18 nodes
        sample_1 = random.sample(range(2400), 18)
        num_byzantine_nodes = sum(1 for node in sample_1 if node < byzantine_nodes)
        if  num_byzantine_nodes > 1:
            sample_2 = random.sample(range(2400), 29)
            num_byzantine_nodes = sum(1 for node in sample_2 if node < byzantine_nodes)
            if num_byzantine_nodes > 5:
                sample_3 = random.sample(range(2400), 38)
                num_byzantine_nodes = sum(1 for node in sample_3 if node < byzantine_nodes)
                if num_byzantine_nodes > 9:
                    sample_4 = random.sample(range(2400), 46)
                    num_byzantine_nodes = sum(1 for node in sample_4 if node < byzantine_nodes)
                    if num_byzantine_nodes > 15:
                        results.append(221)
                    else:
                        results.append(46)
                else:
                    results.append(38)
            else:
                results.append(29)
        else:
            results.append(18)
    # Calculate the average number of records
    return sum(results) / len(results)

gresult = []
goverlap = []
for n in range(50):
    results = []
    for i in range(1000):
        # Run the simulation for 'n' and obtain the average number of records
        avg_records = run_simulation(n)
        results.append(avg_records)
    # Calculate the average results over multiple runs
    avg_results = sum(results) / len(results)
    gresult.append(avg_results)
    print(f"n={n}: {avg_results}")
    # Calculate the overlap time by dividing the average results by 16
    print(f"overlap time: {avg_results/15}")
    goverlap.append(avg_results / 15)
```

```python
import matplotlib.pyplot as plt
# Data
x = range(34)  # x-values
y = goverlap[0:34] # y-values

# Plotting the line graph
plt.plot(x, y)

# Adding labels and title
plt.xlabel('Adversary ratio(%)')
plt.ylabel('Overlap times')
plt.savefig('./fig/Gearbox_Overlap.pdf', format='pdf', transparent=True,bbox_inches='tight')
# Displaying the plot
plt.show()
```
